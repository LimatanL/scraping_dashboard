```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup

def scrape_carsome(page_number):
    url = f"https://www.carsome.id/beli-mobil-bekas?pageNo={page_number}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36"
    }

    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        car_items = soup.find_all("a", class_="mod-b-card__title")
        
        with open(f"carsome_page_{page_number}.csv", "w", newline="", encoding="utf-8") as csv_file:
            writer = csv.writer(csv_file)
            writer.writerow(["Year", "Brand", "Model", "Other Info"])  # Menulis header kolom
            
            for car_item in car_items:
                model_text = car_item.get_text(strip=True)
                
                # Memisahkan nilai berdasarkan "\n"
                split_data = model_text.split("\n")
                if len(split_data) >= 4:  # Memastikan bahwa split_data memiliki setidaknya 4 elemen
                    year = split_data[0].strip()
                    brand_model = split_data[1].strip()
                    variant = split_data[2].strip()
                    other_info = split_data[3].strip()
                    writer.writerow([year, brand_model, variant, other_info])  # Menulis baris data ke file CSV
                else:
                    print(f"Data '{model_text}' tidak lengkap, dilewati.")
                    
        print(f"Data berhasil disimpan dalam file 'carsome_page_{page_number}.csv'.")
    else:
        print("Failed to fetch page.")

# Ganti angka 1 dengan nomor halaman yang ingin Anda scrape
page_number = 1
scrape_carsome(page_number)
```
